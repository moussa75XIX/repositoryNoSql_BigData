## Question 0

### Nettoyer les données

```python
df_ks['launched'] = pd.to_datetime(df_ks['launched'], errors='coerce')
df_ks = df_ks.dropna(subset=['launched'])
df_ks.rename(columns={'ID':'_id'}, inplace=True)

l_d =df_ks.to_dict(orient='records')
```

```python
df_ks.head()
```

### Importer les données

```python
records = df_ks.to_dict(orient='records')
collection.insert_many(records)
```

## Question 1  

```python
# fontionne
cur = collection.aggregate([{"$group":{"_id": "$name", "pledged" :{"$sum":"$pledged"}}}, {"$sort": {"pledged":-1}}, {"$limit": 5}])
list(cur)
```

## Question 2

```python
# fonctionne
cur = collection.find({"state": "successful"})
# list(cur)
cur.count()
```

## Question 3

```python
# fonctionne
cur = collection.aggregate([{"$group" : {"_id" : "$category", "nb_projects" : {"$sum" : 1}}}])
list(cur)
```

```python
# fonctionne
cursor = collection.find({"$and":[{"goal":{"$gte":200000}}]})
list(cursor)
```
